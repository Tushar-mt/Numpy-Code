```python
import matplotlib.pyplot as plt 
import numpy as np

student = ["Tushar", "Aradhya", "Bro", "Kumar"]
marks = [99, 78, 69, 87]

plt.xlabel("Student")
plt.ylabel("Marks")
plt.title("Student Marks")

plt.bar(student, marks, width = 0.2, color = "red")
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np

x1 = np.array([2, 5, 4, 8, 4, 7, 9, 1, 5, 6])
y1 = np.array([22, 66, 77, 99, 44, 77, 22, 66, 11, 77])

x2 = np.array([9, 4, 7, 3, 5, 3, 7, 1, 3, 8])
y2 = np.array([99, 67, 87, 45, 82, 29, 13, 76, 78, 49])

plt.scatter(x1, y1, color = "green", alpha = 0.9, s = 99, label = "Class A")
plt.scatter(x2, y2, color = "red", alpha = 0.9, s = 99, label = "Class B")

plt.xlabel("Hours Studied")
plt.ylabel("Grades")
plt.title("Scatter Plot")

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np

x = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
y = [2, 5, 7, 4, 8, 3, 6]

plt.xlabel("Days")
plt.ylabel("Numbers")
plt.title("Line Plot")

plt.plot(x, y, color = "red", linewidth = 5, marker = 'o', markerfacecolor = "blue", markersize = 12, linestyle = 'dashed')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Basic Histogram')

plt.hist(data, bins=30, color='skyblue', edgecolor='black')

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data = np.random.randn(1000)

plt.xlabel('Values')
plt.ylabel('Density')
plt.title('Customized Histogram with Density Plot')

sns.histplot(data, bins=30, kde=True, color='black', edgecolor='red')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(10)
d = np.random.normal(100, 20, 200)

plt.boxplot(d)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(10)

d_1 = np.random.normal(100, 10, 200)
d_2 = np.random.normal(90, 20, 200)
d_3 = np.random.normal(80, 30, 200)
d_4 = np.random.normal(70, 40, 200)
d = [d_1, d_2, d_3, d_4]

fig = plt.figure(figsize =(10, 7))
ax = fig.add_axes([0, 0, 1, 1])
bp = ax.boxplot(d)

plt.show()
```
